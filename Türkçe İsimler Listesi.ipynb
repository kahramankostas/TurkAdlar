

# https://turkadlar.com/?s=tum-adlar  adresindeki yaklaşık 15000 ismi, cinsiyet ve kullanım sıklığı bilgileriyle birlikte CSV formatında kaydeder. 



# 1992'den önce doğmuş olup 2009'da hayatta olan TC vatandaşlarının ad sıklığı sıralamasıdır.

```python
import urllib
import requests
import pandas as pd
```

```python
former=""
filename="birleşik.csv"
import codecs
cinsiyet={"c=1&ad":"K","c=2&ad":"E"}
dosya =codecs.open(filename, "w", "utf-8")
dosya.write("Cinsiyet,İsim,Taşıyan Kişi Sayısı\n")
for i in range (1,84):
    site="https://turkadlar.com/?s=tum-adlar&kad=&ead=&sh="+str(i)+"&sh2="+str(i)
    temp=urllib.request.urlopen(site)
    sayfa = temp.read().decode("'utf-8'")
    
    for c in cinsiyet:
        while True:
                baş=sayfa.find(c)
                sayfa=sayfa[baş+7:]
                son=sayfa.find("kişi")
                temp=sayfa[0:son]
                temp=temp.replace("</a","")
                a=temp.split(">")
                try:
                    line=cinsiyet[c]+","+a[1]+","+a[4]+"\n"
                    
                except:
                    print("------------------------",i,"--------------------------------")
                    break
                if line!=former: 
                    if "http" not in line:
                        dosya.write(line)
                        print(line[:-1])
                        former=line
dosya.close()
```
